```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('pos_area.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,[1,2]].values
y = dataset.iloc[:,3].values
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
#import the k means cluster
from sklearn.cluster import KMeans

arr = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=789, n_init=10, random_state=0)
  kmeans.fit(X)
  arr.append(kmeans.inertia_)

plt.plot(range(1,11), arr)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
#categorize the data using optimum number of clusters
kmeans = KMeans(n_clusters=7, init='k-means++', max_iter=789, n_init=10, random_state=0)
predict_y = kmeans.fit_predict(X)

plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=788,c='red')
```

# Hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage, fcluster

Z2 = linkage(dataset, method='ward', metric='euclidean')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize=(13, 12))
dendrogram(
    Z2,
    orientation='top',
    distance_sort='descending',
    show_leaf_counts=False,
    leaf_font_size=16
)
plt.show()
```
